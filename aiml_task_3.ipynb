```python
import pandas as pd
df = pd.read_csv("cricket_players.csv")
print(df.head(10))
```

import matplotlib.pyplot as plt
plt.hist(df['Matches'], bins =5 , (0,500) , edgecolor = "black")
plt.title("histogram")
plt.show()

```python
import matplotlib.pyplot as plt
plt.hist(df['Matches'], bins=3 , color = "green" , edgecolor = "white" , orientation = "horizontal" , rwidth = 0.6 , align = "mid")
plt.title("Histogram of matches played")
plt.xlabel("freq")
plt.ylabel("matches")
plt.show()
```

```python
import matplotlib.pyplot as plt
sizes=[100]
plt.scatter(df['Matches'] , df['Runs'] , color = "red" , edgecolor = "yellow" , s=sizes)
plt.title("scatterplot - matches v runs")
plt.xlabel("matches")
plt.ylabel("runs")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
x = df.pivot_table(index = "Name" , values = ["Matches" , "Centuries" , "Age"])
plt.figure(figsize=(5,5))
sns.heatmap(x , annot = True , fmt = 'g' ,vmin=0, vmax=600)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=df["Runs"], color="orange")
sns.swarmplot(x=df["Runs"], color="black", size=6)
plt.title("boxplot of runs scored")
plt.show()

```
